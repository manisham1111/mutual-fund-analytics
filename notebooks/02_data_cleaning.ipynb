```python
fund_master.drop_duplicates(inplace=True)
fund_master = fund_master.apply(
    lambda col: col.str.strip() if col.dtype == "object" else col
)
fund_master.ffill(inplace=True)
if "amfi_code" in fund_master.columns:
    fund_master = fund_master[
        fund_master["amfi_code"].notna()
    ]
if "amfi_code" in fund_master.columns:
    fund_master["amfi_code"] = (
        fund_master["amfi_code"]
        .astype(int)
    )
```

```python
nav_history["date"] = pd.to_datetime(
    nav_history["date"]
)
nav_history = nav_history.sort_values(
    ["amfi_code","date"]
)
nav_history["nav"] = (
    nav_history.groupby("amfi_code")["nav"]
    .ffill()
)
nav_history.drop_duplicates(inplace=True)
nav_history = nav_history[
    nav_history["nav"] > 0
]
```

```python
print(transactions.columns.tolist())
```

```python
transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"]
)

transactions["transaction_type"] = (
    transactions["transaction_type"]
    .str.upper()
    .str.strip()
)

transactions = transactions[
    transactions["amount_inr"] > 0
]

transactions.drop_duplicates(
    inplace=True
)

print("Investor Transactions cleaned successfully!")
```

```python
print(performance.columns.tolist())
```

```python
# Convert return columns to numeric

performance["return_1yr_pct"] = pd.to_numeric(
    performance["return_1yr_pct"],
    errors="coerce"
)

performance["return_3yr_pct"] = pd.to_numeric(
    performance["return_3yr_pct"],
    errors="coerce"
)

performance["return_5yr_pct"] = pd.to_numeric(
    performance["return_5yr_pct"],
    errors="coerce"
)

# Convert expense ratio

performance["expense_ratio_pct"] = pd.to_numeric(
    performance["expense_ratio_pct"],
    errors="coerce"
)

# Keep valid expense ratios

performance = performance[
    (performance["expense_ratio_pct"] >= 0.1)
    &
    (performance["expense_ratio_pct"] <= 2.5)
]

# Remove duplicates

performance.drop_duplicates(inplace=True)

print("Performance dataset cleaned successfully!")
```

```python
print(sip.columns.tolist())
```

```python
sip["month"] = pd.to_datetime(
    sip["month"]
)

sip.drop_duplicates(
    inplace=True
)

sip["sip_inflow_crore"] = sip[
    "sip_inflow_crore"
].ffill()

sip["active_sip_accounts_crore"] = sip[
    "active_sip_accounts_crore"
].ffill()

sip["new_sip_accounts_lakh"] = sip[
    "new_sip_accounts_lakh"
].ffill()

sip["sip_aum_lakh_crore"] = sip[
    "sip_aum_lakh_crore"
].ffill()

sip["yoy_growth_pct"] = sip[
    "yoy_growth_pct"
].ffill()

print("SIP dataset cleaned successfully!")
```

```python
print(category.columns.tolist())
```

```python
category["month"] = pd.to_datetime(
    category["month"]
)

category.drop_duplicates(
    inplace=True
)

category["net_inflow_crore"] = category[
    "net_inflow_crore"
].fillna(0)

print("Category Inflows dataset cleaned successfully!")
```

```python
print(folio.columns.tolist())
```

```python
folio["month"] = pd.to_datetime(
    folio["month"]
)

folio.drop_duplicates(
    inplace=True
)

folio["total_folios_crore"] = folio[
    "total_folios_crore"
].ffill()

folio["equity_folios_crore"] = folio[
    "equity_folios_crore"
].ffill()

folio["debt_folios_crore"] = folio[
    "debt_folios_crore"
].ffill()

folio["hybrid_folios_crore"] = folio[
    "hybrid_folios_crore"
].ffill()

folio["others_folios_crore"] = folio[
    "others_folios_crore"
].ffill()

print("Folio dataset cleaned successfully!")
```

```python
print(holdings.columns.tolist())
print(benchmark.columns.tolist())
```

```python
holdings["portfolio_date"] = pd.to_datetime(
    holdings["portfolio_date"]
)

holdings.drop_duplicates(
    inplace=True
)

holdings["weight_pct"] = holdings[
    "weight_pct"
].fillna(0)

print("Portfolio Holdings cleaned successfully!")
```

```python
benchmark["date"] = pd.to_datetime(
    benchmark["date"]
)

benchmark.sort_values(
    "date",
    inplace=True
)

benchmark.drop_duplicates(
    inplace=True
)

benchmark["close_value"] = benchmark[
    "close_value"
].ffill()

print("Benchmark dataset cleaned successfully!")
```
